```python
import numpy as np

from astropy.table import Table

data = Table.read("Planet_Lightcurve.fits", format='fits')
time = data['time [days]']
flux = data['flux']
```

```python
#temp cleaned
flux_clean = flux + np.random.normal(0,0.1,len(time))

std_total = np.std(flux_clean-flux)
print(std_total)
#0.04 is temp before cleaned data
in_transit = (time > -0.04) & (time < 0.04)
out_transit = ~in_transit
N_in = np.sum(in_transit)
N_out = np.sum(out_transit)
std_in = std_total/np.sqrt(N_in)
std_out = std_total/np.sqrt(N_out)
std_depth = np.sqrt(std_in**2 + std_out**2)
print(f"The total uncertainty of the measurement of depth is: {std_depth}")
print(N_in)
print(N_out)
```
